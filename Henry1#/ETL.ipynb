-Se calcula 'return' dividiendo la columna 'revenue' entre la columna 'budget'.
-Se utiliza np.where() para evitar divisiones por cero. Si 'budget' es igual a 0, se asigna 0 al 'return' correspondiente.
-Se utiliza fillna() para rellenar los valores faltantes (NaN) en la columna 'return' con 0.

```python
movies_df['budget'] = pd.to_numeric(movies_df['budget'], errors='coerce')

movies_df['return'] = movies_df['revenue'] / movies_df['budget']
movies_df['return'] = np.where(movies_df['budget'] != 0, movies_df['revenue'] / movies_df['budget'], 0)
movies_df['return'] = movies_df['return'].fillna(0)
```

-Creamos una funcion que nos ayude de una forma mas rapida el desanidado de columnas

```python

def desanidado_columnas(val):
    try:
        if pd.isnull(val) or isinstance(val, bool):  # Si el valor es NaN o un valor booleano
            return []  # Retorna una lista vacía
        elif isinstance(val, str):  # Si el valor es una cadena
            data = ast.literal_eval(val)  # Evaluar la cadena como una expresión literal
            if isinstance(data, list):  # Si la evaluación es una lista
                return [item['name'] for item in data]  # Extraer el valor de 'name' de cada elemento de la lista y retornarlos en una lista
            else:
                return []  # Si no es una lista, retornar una lista vacía
        else:
            return []  # Si no es una cadena, retornar una lista vacía
    except (SyntaxError, ValueError):  # Manejar errores de sintaxis y valor
        return []  # En caso de error, retornar una lista vacía

movies_df['production_companies'] = movies_df['production_companies'].apply(desanidado_columnas)
movies_df['production_countries'] = movies_df['production_countries'].apply(desanidado_columnas)
movies_df['spoken_languages'] = movies_df['spoken_languages'].apply(desanidado_columnas)

movies_df
```

Ahora vamos con el ETL del dataframe "credits_df"

Estas líneas de código utilizan el método apply() de Pandas para aplicar la función desanidado_columnas a cada elemento de las columnas 'cast' y 'crew' del DataFrame credits_df. La función desanidado_columnas procesará cada valor de la columna según la lógica definida en la función, devolviendo una lista resultante.

```python
credits_df['cast'] = credits_df['cast'].apply(desanidado_columnas)
credits_df['crew'] = credits_df['crew'].apply(desanidado_columnas)
```
